```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl 
import math
```

```python
xlsx_file = 'Data.xlsx'

# header=None: primeira linha vira dados, não nomes das colunas
df = pd.read_excel(xlsx_file, header=None)

print(df)
```

```python
tempDf = pd.Series(df.values.flatten()) 
tempDf = tempDf.sort_values()

print(tempDf)
```

```python

size = len(tempDf)

k = int(np.ceil(np.sqrt(size)))

H = tempDf.max() - tempDf.min()

h = math.ceil(H / k)


amplitude = tempDf.max() - tempDf.min()

classes = []

start = int(tempDf.min())

while start < tempDf.max():
    
    temp = start
    start = start + k
    classes.append([temp , start])
    
    
    
final = pd.DataFrame(classes)

final.columns = ['min','max']

print(final)

```

```python
bins = list( final['min']) + [final['max'].iloc[-1]]


fi = pd.cut(tempDf, bins=bins, right=False)

frequencia = fi.value_counts().sort_index()

final['fi'] = frequencia.values

final['Xm'] = (final['min'] + final['max']-1) / 2
final.loc[final.index[-1], 'Xm'] = (final['min'].iloc[-1] + final['max'].iloc[-1]) / 2

print(final)
```

```python

final['Xm'] = (final['min'] + final['max'] -1) / 2

final.loc[final.index[-1], 'Xm'] = (final['min'].iloc[-1] + final['max'].iloc[-1]) / 2

print(final)
```

```python
fas = []
fa = 0
for i in range(len(final)):
    fa += final['fi'].iloc[i]
    fas.append(fa)

final['fa'] = fas

print(final)


```

```python
from turtle import pos

from pandas.core.indexing import length_of_indexer


media = 0

for i in range(len(final)):
    media += final['Xm'].iloc[i] * final['fi'].iloc[i]

media = media / len(tempDf)



print("media:", media)

posicao = 0
n = len(tempDf)
if n % 2 == 0:
    posicao = ((n/2) + (n/2 + 1)) / 2
else:
    posicao = (n+1)/2

limiteInferior = 0

fi = 0

for i, frequenciaAcumulada in enumerate(final['fa']):
   if frequenciaAcumulada >= posicao:
        classeMediana = final['min'].iloc[i]
   
        frequenciaAnterior = final['fa'].iloc[i-1] if i > 0 else 0
    
        fi = final['fi'].iloc[i]
    
        limiteInferior = final['min'].iloc[i]
    
        break


mediana = limiteInferior + ((posicao - frequenciaAnterior) / fi) * h

print("mediana:", mediana)
```

```python
variancia = 0

for i in range(len(final)):
    variancia += ((final['Xm'].iloc[i] - media) ** 2) 

variancia = variancia / len(tempDf)

print("variancia:", variancia)

desvioPadrao = math.sqrt(variancia)

print("desvio padrao:", desvioPadrao)

coeficienteVariacao = (desvioPadrao / media) * 100

print("coeficiente de variacao:", coeficienteVariacao)
```
